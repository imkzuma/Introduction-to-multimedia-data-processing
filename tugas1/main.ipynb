```python
import pandas as pd

df = pd.read_excel("dataset.xlsx")

# filter data sesuai absensi
# absen 5 berarti dari 41-50
df = df.iloc[40:50]
df
```

```python
# case folding 
def case_folding(text):
    import re
    # mengubah text menjadi huruf kecil
    # menghapus angka
    # menghapus tanda baca
    # menghapus whitespace
    for i in range(len(text)):
        text = re.sub(r'[0-9]', ' ', text)
        text = re.sub(r'[^\w\s]', ' ', text)
        text = text.lower()
        text = text.strip()
        text = text.replace('\n', '')
    
    return text


text = df['Reviews'].values.tolist()
for i in text : print(i)

print()

text = [case_folding(i) for i in text]
for x in text : print(x)
```

```python
import nltk

def tokenization(text):
    # mengubah text menjadi token
    tokens = list()
    for i in text:
        tokens.append(nltk.word_tokenize(i))

    return tokens

tokens = tokenization(text)

for token in tokens : print(token)
```

```python
singkatan_dict = {
    'tp': 'tapi',
    'kyknya': 'kayaknya',
    'klw': 'kalau',
    'gk': 'tidak',
    'ga' : 'tidak',
    'pdhl': 'padahal',
    'yg': 'yang',
    'ny': 'nya',
    'tpii': 'tapi',
    'tpi' : 'tapi',
    'kyak': 'kayak',
    'kaka' : 'kakak',
    'sma' : 'sama',
    'coklat' : 'cokelat',
    'bahanya' : 'bahannya',
    'ko' : 'kok',
    'y' : '',
    'hargany' : 'harganya',

}

alay_dict = {
    'bagusss': 'bagus',
    'baguss' : 'bagus',
    'bngaaaatt': 'banget',
    'bangetttt': 'banget',
    'nyata': 'beneran',
    'jg': 'juga',
    'makas': 'makasih',
    'nyaesel': 'nyesel',
    'tapii': 'tapi',
    'nyaa' : 'nya',
    'kemejanyaa' : 'kemejanya',
    'makasi' : "terima kasih",
    'cuman' : 'cuma',
    'gapapa' : 'tidak apa-apa',
    'ga' : 'tidak',
    'hehe' : ''
}

# convert text ke list tiap kata
def text_to_list(text):
    listtext = list()
    for i in text.split():
        listtext.append(i)
    
    return listtext

def alay_singkatan_removal(text):
    # mengubah singkatan menjadi kata dasar
    # mengubah kata alay menjadi kata dasar
    for i in range(len(text)):
        text[i] = text_to_list(text[i])
        for j in range(len(text[i])):
            if text[i][j] in singkatan_dict:
                text[i][j] = singkatan_dict[text[i][j]]
            elif text[i][j] in alay_dict:
                text[i][j] = alay_dict[text[i][j]]
        
    
    return text

text = alay_singkatan_removal(text)

for i in text : print(i)
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

def stopword_removal(text):
    # menghapus stopword
    factory = StopWordRemoverFactory()
    stopwords = factory.create_stop_word_remover()

    stop = list()
    for i in text:
        stop.append(stopwords.remove(' '.join(i)))
    
    return stop

stop = stopword_removal(text)
tokens = tokenization(stop)

for token in tokens : print(token)
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stemming(text):
    # mengubah kata menjadi kata dasar
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()

    stem = list()
    for i in text:
        stem.append(stemmer.stem(i))

    return stem

stem = stemming(stop)
for i in stem : print(i)

tokens = tokenization(stem)
for token in tokens : print(token)


```

```python
from sklearn.feature_extraction.text import CountVectorizer

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)

#term document matrix one hot encoding
def one_hot_encoding(text):
    cv = CountVectorizer(binary=True)
    cv.fit(text)
    return cv.transform(text).toarray()

one_hot = pd.DataFrame(one_hot_encoding(stem))
one_hot
```

```python
# term document matrix basic bow
def basic_bow(text):
    cv = CountVectorizer(binary=False)
    cv.fit(text)
    return cv.transform(text).toarray()

basic_bow = pd.DataFrame(basic_bow(stem))
basic_bow
```

```python
#term document matrix tfidf bow
from sklearn.feature_extraction.text import TfidfVectorizer

def tfidf_bow(text):
    cv = TfidfVectorizer()
    cv.fit(text)
    return cv.transform(text).toarray()

tfidf_bow = pd.DataFrame(tfidf_bow(stem))
tfidf_bow
```

```python
# tampilkan kata yang paling sering muncul
def most_common_words(text):
    from collections import Counter

    words = list()
    for i in text:
        for j in i.split():
            words.append(j)
    
    return Counter(words).most_common(10)

common_words = most_common_words(stem)
```

```python
from matplotlib import pyplot as plt

def plot_common_words(common_words):
    # plot kata yang paling sering muncul
    words = list()
    count = list()
    for i in common_words:
        words.append(i[0])
        count.append(i[1])
    
    plt.bar(words, count)
    plt.title('Most Common Words')
    plt.xlabel('Words')
    plt.ylabel('Count')

    plt.show()

plot_common_words(common_words)
```
