# Import Library untuk membaca dataset

```python
import pandas as pd

df = pd.read_excel("dataset.xlsx")

# filter data sesuai absensi
# absen 5 berarti dari 41-50
df = df.iloc[40:50]
df
```

# Fungsi Case Folding

```python
# case folding menggunakan regex
def case_folding(text):
    import re
    # mengubah text menjadi huruf kecil
    # menghapus angka
    # menghapus tanda baca
    # menghapus whitespace
    for i in range(len(text)):
        text = re.sub(r'[0-9]', ' ', text)
        text = re.sub(r'[^\w\s]', ' ', text)
        text = re.sub(r'\s+', ' ', text)
        text = text.lower()
        text = text.strip()
        text = text.replace('\n', '')
    
    return text

# mengubah data menjadi list
text = df['Reviews'].values.tolist()

# mengubah data dengan case folding
text = [case_folding(i) for i in text]

# mengubah data menjadi dataframe
caseFoldingDF = pd.DataFrame(text)
caseFoldingDF
```

# Membuat fungsi tokenisasi

```python
import nltk

# membuat fungsi untuk tokenization menggunakan nltk
def tokenization(text):
    # mengubah text menjadi token
    tokens = list()
    for i in text:
        tokens.append(nltk.word_tokenize(i))

    return tokens

# membuat variable untuk menampung hasil tokenization
tokens = tokenization(text)
for token in tokens : print(token)
```

```python
# kamus singkatan dan alay
singkatan_dict = {
    'tp': 'tapi',
    'kyknya': 'kayaknya',
    'klw': 'kalau',
    'gk': 'tidak',
    'ga' : 'tidak',
    'pdhl': 'padahal',
    'yg': 'yang',
    'ny': 'nya',
    'tpii': 'tapi',
    'tpi' : 'tapi',
    'kyak': 'kayak',
    'kaka' : 'kakak',
    'sma' : 'sama',
    'coklat' : 'cokelat',
    'bahanya' : 'bahannya',
    'ko' : '',
    'y' : '',
    'hargany' : 'harganya',
    'kaliny' : 'kalinya',
    'pengiriman' : 'pengirimannya',
    'nya' : '',
}

alay_dict = {
    'bagusss': 'bagus',
    'baguss' : 'bagus',
    'bngaaaatt': 'banget',
    'bangetttt': 'banget',
    'bgt' : 'banget',
    'dsni' : 'disini',
    'jg': 'juga',
    'makas': 'makasih',
    'nyaesel': 'nyesel',
    'tapii': 'tapi',
    'nyaa' : 'nya',
    'kemejanyaa' : 'kemejanya',
    'makasi' : "terima kasih",
    'cuman' : 'cuma',
    'gapapa' : 'tidak apa-apa',
    'ga' : 'tidak',
    'hehe' : '',
    'tebel' : 'tebal',
    'di' : ''
}

# convert text ke list tiap kata
def text_to_list(text):
    listtext = list()
    for i in text.split():
        listtext.append(i)
    
    return listtext

def alay_singkatan_removal(text):
    # mengubah singkatan menjadi kata dasar
    # mengubah kata alay menjadi kata dasar
    for i in range(len(text)):
        text[i] = text_to_list(text[i])
        for j in range(len(text[i])):
            if text[i][j] in singkatan_dict:
                text[i][j] = singkatan_dict[text[i][j]]
            elif text[i][j] in alay_dict:
                text[i][j] = alay_dict[text[i][j]]
                
    return text

text = alay_singkatan_removal(text)

for i in text : print(i)
```

# Membuat fungsi stopword removal

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

def stopword_removal(text):
    # menghapus stopword
    factory = StopWordRemoverFactory()
    stopwords = factory.create_stop_word_remover()

    stop = list()
    for i in text:
        stop.append(stopwords.remove(' '.join(i)))
    
    return stop

stop = stopword_removal(text)
tokens = tokenization(stop)

for token in tokens : print(token)
```

# Membuat fungsi stemming 
### fungsi ini mengubah kata menjadi kata dasar

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stemming(text):
    # mengubah kata menjadi kata dasar
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()

    stem = list()
    for i in text:
        stem.append(stemmer.stem(i))

    return stem

stem = stemming(stop)

showStem = pd.DataFrame(stem)
stem = tokenization(stem)

showStem
```

# Membuat fungsi one hot encoding

```python
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)

def create_word_list(text):
    wordList = list()
    for key in text:
        for word in key:
            if word not in wordList : wordList.append(word)

    return sorted(wordList)

def one_hot_encoding(text):
    wordList = create_word_list(text)
    
    ohe = np.zeros((len(stem), len(wordList)) , dtype=int)

    for i, key in enumerate(stem):
        for word in wordList:
            if word in key:
                ohe[i][wordList.index(word)] = 1

    ohe = pd.DataFrame(ohe, columns=wordList, index=["Doc {}".format(i) for i in range(len(text))])

    return ohe

ohe = one_hot_encoding(stem)
ohe.T

```

# Membuat fungsi basic bow

```python
# term document matrix basic bow

def basic_bow(text):
    wordList = create_word_list(text)

    bow = np.zeros((len(stem), len(wordList)) , dtype=int)

    for i, key in enumerate(stem):
        for word in key:
            bow[i][wordList.index(word)] += 1
    
    bow = pd.DataFrame(bow, columns=wordList, index=["Doc {}".format(i) for i in range(len(text))])

    return bow

bow = basic_bow(stem)
bow.T
```

# Membuat fungsi tfidf_bow

```python
# tfidf
from math import log10


def tfidf(text):
    wordList = create_word_list(text)

    matrix = np.zeros((len(stem), len(wordList)) , dtype=float)
    ohedf = one_hot_encoding(stem)
    bowdf = basic_bow(stem)

    dft = dict()

    for word in wordList:
        for index in stem:
            if word not in dft:
                dft[word] = 0
            if word in index:
                dft[word] += 1

    for i, key in enumerate(stem):
        for word in key:
            matrix[i][wordList.index(word)] = (1 + np.log10(bowdf.iloc[i][wordList.index(word)])) * np.log10(len(stem) / dft[word])

    matrix = pd.DataFrame(matrix, columns=wordList, index=["Doc {}".format(i) for i in range(len(text))])

    return matrix

tfidf = tfidf(stem)
tfidf.T
```

# Membuat fungsi untuk menghitung banyaknya kata yang muncul dalam dokumen

```python
# tampilkan kata yang paling sering muncul
def most_common_word(text):
    wordList = create_word_list(text)

    bowdf = basic_bow(stem)

    df = pd.DataFrame(bowdf.sum(), columns=['Jumlah'])
    df = df.sort_values(by='Jumlah', ascending=False)
    df = df.reset_index()
    df = df.rename(columns={'index': 'Kata'})

    #tamplikan 10 kata yang paling sering muncul
    df = df.head(10)

    return df

comm_word = most_common_word(stem)
comm_word
```
